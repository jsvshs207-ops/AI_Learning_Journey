##### 二、Plot(x, y)    点线图

plot 图是Matplotlib中最常用的图表类型之一，用于绘制折线图。它将数据点连接起来，形成平滑的曲线，以显示数据的趋势和变化。



```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery')

# make data
x = np.linspace(0, 10, 100)
y = 4 + 2 * np.sin(2 * x)

# plot
fig, ax = plt.subplots()

"""
创建图形对象和子图对象的目的是为了更灵活地控制和管理图形的各个方面。

直接调用函数画图是可以的,例如使用plt.scatter()函数绘制散点图。但是，当需要对图形进行更多的自定义和控制时，使用图形对象和子图对象会更加方便。

通过创建图形对象和子图对象，可以实现以下功能：
- 在同一个图形中绘制多个子图，每个子图可以有不同的布局和样式。
- 对图形进行更多的自定义设置，如设置图形的大小、标题、坐标轴标签等。
- 在子图上绘制多个图形元素，如散点图、线图、柱状图等。
- 对子图进行更多的自定义设置，如设置坐标轴范围、刻度、标签等。

此外，通过图形对象和子图对象，还可以方便地保存图形为文件，或者在同一个图形中显示多个子图。

因此，创建图形对象和子图对象可以提供更多的灵活性和控制力，使得绘图过程更加方便和可定制。 
"""

ax.plot(x, y, linewidth=2.0)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

##### 三、Scatter（x, y）   散点图
scatter 图是Matplotlib中的一种图表类型，用于绘制散点图。它将数据点以离散的方式显示在二维平面上，每个数据点由一对坐标表示。



```python
# 导入matplotlib.pyplot和numpy库
import matplotlib.pyplot as plt
import numpy as np

# 使用'_mpl-gallery'样式
plt.style.use('_mpl-gallery')

# 生成数据
np.random.seed(3)  # 设置随机种子，保证每次运行结果一致
x = 4 + np.random.normal(0, 2, 24)  # 生成24个以4为中心的正态分布随机数
y = 4 + np.random.normal(0, 2, len(x))  # 生成与x长度相同的以4为中心的正态分布随机数

# 设置散点的大小和颜色
sizes = np.random.uniform(15, 80, len(x))  # 生成与x长度相同的在15到80之间的均匀分布随机数，作为散点的大小
colors = np.random.uniform(15, 80, len(x))  # 生成与x长度相同的在15到80之间的均匀分布随机数，作为散点的颜色

# 创建图形和子图
fig, ax = plt.subplots()

# 在子图上绘制散点图，x和y为坐标，sizes为大小，colors为颜色，vmin和vmax设置颜色的范围
ax.scatter(x, y, s=sizes, c=colors, vmin=0, vmax=100)

# 设置x和y轴的范围和刻度
ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

# 显示图形
plt.show()
```

##### 四、bar(x,height)   柱状图
bar 图是Matplotlib中常用的一种图表类型，用于绘制柱状图。它通过多个垂直或水平的柱子来表示数据的大小或数量。



```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('_mpl-gallery')

# make data:
x = 0.5 + np.arange(8)
y = [4.8, 5.5, 3.5, 4.6, 6.5, 6.6, 2.6, 3.0]

# plot
fig, ax = plt.subplots()

ax.bar(x, y, width=1, edgecolor="white", linewidth=0.7)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

##### 五、stem(x,y)  棍图
"stem"图是一种用于表示离散数据的图表类型。它通过在坐标轴上垂直放置的线段或者称为“棍”的形式，将每个数据点与其对应的数值关联起来。

stem图常用于以下情景：

* 表示序列数据的变化趋势：stem图可以将序列数据按照时间、位置等信息绘制出来，以展示数据的变化趋势和特征。

* 表示离散数据的分布：当数据是离散的、非连续的时候，stem图可以清晰地显示每个数据点的位置和数值。

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery')

# make data
x = 0.5 + np.arange(8)
y = [4.8, 5.5, 3.5, 4.6, 6.5, 6.6, 2.6, 3.0]

# plot
fig, ax = plt.subplots()

ax.stem(x, y)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

##### 六、step(x,y) 阶梯图
"step"图是一种用于表示离散数据或分段函数的图表类型。它通过水平和垂直方向上的线段连接离散的数据点或分段函数的值，形成一个阶梯状的图形。

Step图常用于以下情景：

* 表示离散数据的变化趋势：当数据是离散的、非连续的时候，step图可以清晰地显示每个数据点的位置和数值，并且通过阶梯状的线段来展示数据的变化趋势。

* 表示分段函数的曲线：如果有一个函数在不同的区间内具有不同的定义或规则，step图可以用来表示这个分段函数的值，并且在每个区间内形成阶梯状的曲线。



```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery')

# make data
x = 0.5 + np.arange(8)
y = [4.8, 5.5, 3.5, 4.6, 6.5, 6.6, 2.6, 3.0]

# plot
fig, ax = plt.subplots()

ax.step(x, y, linewidth=2.5)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

##### 七、误差棒 fill_between(x,y1,y2)
fill_between 是Matplotlib中的一个函数，用于在两条曲线之间填充颜色。它可以用来表示两个数据序列之间的区域或者标识出某个数据序列所占据的范围。

fill_between 函数常用于以下情景：

* 表示误差范围：当有两个数据序列表示上下界限时，可以使用 fill_between 函数在这两条曲线之间填充颜色，以表示数据的误差范围。

* 标记特定区域：通过与其他曲线结合使用，可以使用 fill_between 函数标记出某个数据序列所占据的范围，以突出显示数据的特定部分。

在Matplotlib中，你可以使用 plt.fill_between() 函数来创建 fill_between 图形。你需要提供X轴上的数据点位置和相应的两个数据序列作为参数，然后可以根据需要选择不同的样式、颜色等进行定制。

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery')

# make data
np.random.seed(1)
x = np.linspace(0, 8, 16)
y1 = 3 + 4*x/8 + np.random.uniform(0.0, 0.5, len(x))
y2 = 1 + 2*x/8 + np.random.uniform(0.0, 0.5, len(x))

# plot
fig, ax = plt.subplots()

ax.fill_between(x, y1, y2, alpha=.5, linewidth=0)
ax.plot(x, (y1 + y2)/2, linewidth=2)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

##### 八、stackplot(x,y) 堆叠图
stackplot 图是一种用于展示多个数据序列叠加在一起的图表类型。它将多个序列分别表示为不同的区域，并按照给定的顺序将它们垂直叠加在一起。

stackplot 图常用于以下情景：

* 表示组成部分的比例：当需要展示数据中各个组成部分的比例关系时，可以使用 stackplot 图来清晰地显示每个组成部分占总体的比例。

* 展示时间序列的数据变化：如果有多个时间序列数据，可以使用 stackplot 图将它们叠加在一起，以直观地显示它们之间的演变和相对大小。

* 在Matplotlib中，你可以使用 plt.stackplot() 函数来创建 stackplot 图形。你需要提供X轴上的数据点位置和一个由多个数据序列组成的列表作为参数，然后可以根据需要选择不同的样式、颜色和标签等进行定制。

stackplot 图可以根据需要进行堆叠的层数和堆叠顺序，也可以添加透明度以增强可视效果。此外，你还可以使用 labels 参数指定每个数据序列的标签，并使用图例来解释每个区域的含义。

stackplot 图是一种直观清晰地表示多个数据序列叠加关系的可视化方式。它能够帮助我们更好地理解和展示数据的组成部分以及各个部分之间的比例关系。

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery')

# make data
x = np.arange(0, 10, 2)
ay = [1, 1.25, 2, 2.75, 3]
by = [1, 1, 1, 1, 1]
cy = [2, 1, 2, 1, 2]
y = np.vstack([ay, by, cy])

# plot
fig, ax = plt.subplots()

ax.stackplot(x, y)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```
